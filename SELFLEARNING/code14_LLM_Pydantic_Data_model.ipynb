```python
model_name = "gemini-2.0-flash"

from langchain.chat_models import init_chat_model
llm = init_chat_model(model_name, model_provider="google_genai")
```

```python
prompt = "Answer the question in the json format: what is capital of India? with key as 'asnwer'"
response = llm.invoke(prompt)
print(response.content)
```

```python
from pydantic import BaseModel,Field

class responseFormat(BaseModel):
    answer:str = Field(description="answer to the question in one word/phrase")

structured_llm = llm.with_structured_output(responseFormat)

response = structured_llm.invoke(prompt)
```
