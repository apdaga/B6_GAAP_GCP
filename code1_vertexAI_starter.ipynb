```python
from google import genai
import os
client = genai.Client()
```

```python
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="Explain QUantum Computing in simple terms, one line")
print(response.text)
```

```python
from google.genai import types
contents = [types.Content(role='user',parts=[types.Part(text="Explain quantum computing in 1 line")])]
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=contents)
print(response.text)
```
